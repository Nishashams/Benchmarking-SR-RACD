```python
import os
import subprocess

def create_video(input_folder, output_file, fps=30):
    # List all image files in the input folder
    image_files = [f for f in os.listdir(input_folder) if f.endswith('.png')]

    # Sort the image files based on their names
    image_files.sort()

    # Generate the input file list for ffmpeg
    input_files = [os.path.join(input_folder, f) for f in image_files]

    # Run ffmpeg command to create the video
    cmd = [
        'ffmpeg',
        '-framerate', str(fps),
        '-i', os.path.join(input_folder, 'frame_%08d.png'),
        '-c:v', 'libx264',
        '-r', '30',
        '-pix_fmt', 'yuv420p',
        output_file
    ]

    process = subprocess.run(cmd, stderr=subprocess.PIPE)
    stderr = process.stderr.decode('utf-8')

    if process.returncode != 0:
        raise RuntimeError(f"Failed to create video: {stderr}")
    else:
        print("Video created successfully!")

# Set the paths
input_folder = '/content/tmp_frames'
output_file = '/content/drive/MyDrive/dataset/output_videoo.mp4'

# Create the video
create_video(input_folder, output_file)
```

```python
import cv2
from skimage.metrics import peak_signal_noise_ratio, structural_similarity

# Load the original and super-resolved videos
original_video_path = '/content/drive/MyDrive/satellite.mp4'
super_res_video_path = '/content/drive/MyDrive/dataset/output_videoo.mp4'

original_video = cv2.VideoCapture(original_video_path)
super_res_video = cv2.VideoCapture(super_res_video_path)

# Initialize variables to store PSNR and SSIM values
psnr_values = []
ssim_values = []

# Loop through the frames of the videos
while True:
    # Read frames from the videos
    ret_o, frame_o = original_video.read()
    ret_sr, frame_sr = super_res_video.read()

    # Break the loop if any of the videos end
    if not ret_o or not ret_sr:
        break

    # Convert frames to grayscale (if the videos are in color)
    frame_o_gray = cv2.cvtColor(frame_o, cv2.COLOR_BGR2GRAY)
    frame_sr_gray = cv2.cvtColor(frame_sr, cv2.COLOR_BGR2GRAY)

    # Calculate PSNR
    psnr_value = peak_signal_noise_ratio(frame_o_gray, frame_sr_gray)
    psnr_values.append(psnr_value)

    # Calculate SSIM
    ssim_value, _ = structural_similarity(frame_o_gray, frame_sr_gray, full=True)
    ssim_values.append(ssim_value)

# Calculate average PSNR and SSIM values
average_psnr = sum(psnr_values) / len(psnr_values)
average_ssim = sum(ssim_values) / len(ssim_values)

# Print the results
print(f'Average PSNR: {average_psnr:.2f}')
print(f'Average SSIM: {average_ssim:.4f}')

# Release video capture objects
original_video.release()
super_res_video.release()
```
